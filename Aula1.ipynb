# Automação de Sistemas e Processos com Python

### Desafio:

Para controle de custos, todos os dias, seu chefe pede um relatório com todas as compras de mercadorias da empresa.
O seu trabalho, como analista, é enviar um e-mail para ele, assim que começar a trabalhar, com o total gasto, a quantidade de produtos compradas e o preço médio dos produtos.

E-mail do seu chefe: para o nosso exercício, coloque um e-mail seu como sendo o e-mail do seu chefe<br>
Link de acesso ao sistema da empresa: https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
import pyautogui
import time
import pandas as pd
import pyperclip as pclip

#https://pyautogui.readthedocs.io/en/latest/quickstart.html

#pyautogui.click
#pyautogui.write
#pyautogui.press
#pyautowgui.hotkey -> Aperta uma combinação de teclas (Ex: Ctrl + D)

#Passso a Passo

pyautogui.PAUSE = 1 #Espera em todos os comandos

#Passo 1: Entrar no site da empresa
# pyautogui.press("win")
# pyautogui.write("chrome")
# pyautogui.press("enter")

pyautogui.hotkey("ctrl", "t")
pyautogui.write("https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema")
pyautogui.press("enter")

#Espera apenas neste momento
# print(pyautogui.position())

#Passo 2: Fazer Login
pyautogui.click(x=3530, y=464)
pyautogui.write("Meu_login")


pyautogui.click(x=3534, y=580)
pyautogui.write("Minha_senha")

pyautogui.click(x=3520, y=653)

time.sleep(3)

#Passo 3: Exportar a base de dados
pyautogui.click(x=2873, y=506, button="right")
pyautogui.click(x=3040, y=1115)
pyautogui.click(x=3393, y=640)
time.sleep(2)
pyautogui.click(x=3393, y=640)

#Passo 4: Calcular os indicadores 
tabela = pd.read_csv(r"C:\Users\Felipe\Documents\hashtagTreinamentos\Compras.csv", sep=";") 
#sep = separador
#Tem que pegar o caminho do arquivo
#'r' = raw string
display(tabela)

total_gasto = tabela["ValorFinal"].sum()
quantidade_total = tabela["Quantidade"].sum()
preco_medio = total_gasto / quantidade_total

print(total_gasto)
print(quantidade_total)
print(preco_medio)

#Passo 5: Enviar um email 
pyautogui.hotkey("ctrl", "t")
pyautogui.write("https://mail.google.com/mail/u/1/#inbox")
pyautogui.press("enter")
time.sleep(2)
pyautogui.click(x=2504, y=243)
pyautogui.write("feliep8@gmail.com")

pyautogui.press("tab") #Escolhe o destinatario

pyautogui.press("tab") #Passa para o campo assunto
pyautogui.write("Meu Assunto")

pyautogui.press("tab") #Passa para o corpo do email

texto = f"""
Prezados,
Segue o relatório de compras

Total Gasto: R${total_gasto:,.2f} 

Quantidade Total: {quantidade_total:,}

Preço Médio: R${preco_medio:,.2f}

Qualquer duvida, é só chamar!!!
Att., Felipe
"""

pclip.copy(texto)
pyautogui.hotkey("ctrl", "v")

pyautogui.hotkey("ctrl", "enter") #enviar
```
